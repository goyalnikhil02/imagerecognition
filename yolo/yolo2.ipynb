```python
import cv2
```

```python
from darkflow.net.build import TFNet
```

```python
import matplotlib.pyplot as plt
```

```python
% config InlineBakcend.figure_format='svg'
```

```python
options ={
    'model':'C:/Users/g521784/yolo/darkflow-master/cfg/yolo.cfg',
    'load': 'C:/Users/g521784/yolo/darkflow-master/bin/yolo.weights',
    'threshold':0.3    
}
tfnet=TFNet(options)
img=cv2.imread('C:/Users/g521784/test_images/image2.jpeg')
#img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
result=tfnet.return_predict(img)
result
```

```python
for elem in result:
        t1test=(elem['topleft']['x'],elem['topleft']['y'])
        t2test=(elem['bottomright']['x'],elem['bottomright']['y'])
        labeltest=(elem['label'])
        img=cv2.rectangle(img, t1test,t2test, (255,0,0), 5)
        img=cv2.putText(img,labeltest,t1test,cv2.FONT_HERSHEY_COMPLEX,2,(0,0,0),2)

        
plt.imshow(img)
plt.show()
cv2.imwrite('result2.jpg', img)
cv2.destroyAllWindows()
        
#t1=(result[0]['topleft']['x'],result[0]['topleft']['y'])
#t2=(result[0]['bottomright']['x'],result[0]['bottomright']['y'])
#label=result[0]['label']
#print(type(result))
```
